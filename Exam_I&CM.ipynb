```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
raw_data=pd.read_csv('D:\\STUDY\\Semester3\\G-19 Innovation and Complexity Management\\cases.csv')
raw_data.head(10)
```

```python
sns.set(rc={'figure.figsize':(16,8)})

sns.set_style('darkgrid',
              {'axes.facecolor': '1',
               'grid.color': '0.0',
               'figure.facecolor': '0.8'}) 
```

```python
ax = sns.barplot(x = 'year', 
                 y = 'cases', 
                 data = raw_data, 
                 ci = None,
                 hue = 'gender',
                 estimator = sum, 
                 palette = 'rocket_r')

#for p in ax.patches:
#    ax.annotate(format(p.get_height(), '.1f'), 
#                   (p.get_x() + p.get_width() / 2., p.get_height()), 
#                   ha = 'center', va = 'center', 
#                   xytext = (0,0), 
#                   textcoords = 'offset points')

ax.set_title('Total number of new diabetes cases for the last 10 years', size = 20);
ax.set(xlabel = '', ylabel = 'Total cases')
plt.gca().legend().set_title('Gender')

def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        patch.set_width(new_value)

change_width(ax, .28)
```

```python
ax = sns.barplot(x = 'year', 
                 y = 'cases', 
                 data = raw_data, 
                 ci = None, 
                 hue = 'country',
                 estimator = sum, 
                 palette = 'rocket_r')

#for p in ax.patches:
#    ax.annotate(format(p.get_height(), '.1f'), 
#                   (p.get_x() + p.get_width() / 2., p.get_height()), 
#                   ha = 'center', va = 'center', 
#                   xytext = (0,0), 
#                   textcoords = 'offset points')

ax.set_title('Total number of new diabetes cases by countries', size = 20);
ax.set(xlabel = '', ylabel = 'Total cases')
plt.gca().legend().set_title('Countries')           
```
